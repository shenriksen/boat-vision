# Boat Vision — train the buoy/maritime model (free GPU)

Trains a model that detects **buoys, boats, vessels** on the Singapore Maritime dataset.
Runs on a free cloud GPU, no install on your PC.

**Steps:**
1. Menu **Runtime → Change runtime type → Hardware accelerator: GPU → Save**.
2. Menu **Runtime → Run all** (Ctrl+F9).
3. Paste your Roboflow API key when asked.
4. Wait ~30–45 min. At the end it downloads **best.pt**.
5. Put **best.pt** into Boat Vision at: `models\maritime\best.pt` and restart the app.

```python
import torch
print('GPU available:', torch.cuda.is_available(), torch.cuda.get_device_name(0) if torch.cuda.is_available() else '(none - set Runtime->GPU)')
!pip -q install ultralytics roboflow
```

```python
import getpass
from roboflow import Roboflow
key = getpass.getpass('Paste your Roboflow API key and press Enter: ')
rf = Roboflow(api_key=key)
ds = rf.workspace('maritime-cumkb').project('singapore-maritime').version(5).download('yolov8')
print('dataset downloaded to', ds.location)
```

```python
from ultralytics import YOLO
model = YOLO('yolo26s.pt')
model.train(data=ds.location + '/data.yaml', epochs=60, imgsz=640, batch=16, name='maritime_buoy')
```

```python
best = str(model.trainer.best)
print('Best weights:', best)
from google.colab import files
files.download(best)   # downloads best.pt to your computer
```
